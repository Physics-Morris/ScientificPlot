```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import AutoMinorLocator
from scipy.special import legendre

def apply_custom_style(ax):
    # Hide top and right spines
    ax.spines["top"].set_visible(False)
    ax.spines["right"].set_visible(False)

    # Set tick parameters
    ax.tick_params(axis='both', pad=5, labelsize=15, which='both')  # 'which' can be 'major', 'minor', or 'both'

    # Set legend properties
    ax.legend(loc='right', fontsize=13, ncol=1)

    # Set minor ticks
    ax.xaxis.set_minor_locator(AutoMinorLocator(4))
    ax.yaxis.set_minor_locator(AutoMinorLocator(4))

plt.style.use('my.mplstyle')

fig, ax1 = plt.subplots()
apply_custom_style(ax1)

x = np.linspace(-1, 1, 400)
leg_poly = [legendre(i)(x) for i in range(5)]

for i, y in enumerate(leg_poly, 1):
    ax1.plot(x, y, label=f'$P_{{{i-1}}}(x)$')

ax1.set_xlabel('$x$')
ax1.set_ylabel('$y$')
ax1.set_title('Legendre Polynomials')

plt.legend()
plt.tight_layout()
plt.savefig('example_plot.pdf')
plt.show()
```
